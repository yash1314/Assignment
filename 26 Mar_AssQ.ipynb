#### Q1. Explain the difference between simple linear regression and multiple linear regression. Provide an example of each.






Ans: In statistical terms 'Regression' is the way of establishing and determining the relationship between one dependent variable and one or more independent variabels. Such a method is used almost everywhere i.e in finanace, banking, medical, manufacturing, etc. 
Now comes simple linear regression and multiple linear regression.

Simple Linear regression : In Simple linear regression we are trying to find the best fit line while minimizing the sum of squared errors (loss function) to end up with the line/equation which best describe the realtionship between output and input variables. When we are talking about a line we mean a equation of a line which is:

=> y = b0 + b1.x            ----eq(1)

where, y = output variable

       b0 = intercept of line at y

       b1 = slope of a line

       x = input variable.

We use Mean squared error to find most efficient parameters that will create best fit line. Mean squared error is the average difference between the actual target values to the predicted y values.
In simple linear regression we strictly have one input variable and one output variable. 

Multiple Linear regression : whereas Multiple linear regression is almost same as simple linear regression except instead of one input variable we have two or more input variable and only one output variable. 

equation for multiple linear regression is:

y = b0 + b1.x1 + b2.x2 + b3.x3 + ---- +  bn.xn

where, y = output variable

       b0 = intercept of line at y

       b1, b2, b3 -- bn = n - slopes  

       x1, x2, --- xn = n - multiple independent input variables.
       
       
- With multiple linear regression we can explore, explain and predict complex real world phenomena because of multiple input variables. whereas in case of simple linear regression we can only explain and predict relationship between one phenomena to other. 

#### Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in a given dataset?

- The assumptions of linear regression are:

a) There is a linear relationship between the predictors (x) and the outcome (y).

b) Predictors (x) are independent (no correlation) and observed with negligible error.

c) Residual Errors have a mean value of zero.

d) Residual Errors have constant variance.
 
e) Residual Errors are independent from each other and predictors (x).

- Assumption One: Linearity of the Data

We can check the linearity of the data by looking at the Residual vs Fitted plot. Ideally, this plot would not have a pattern where the red line (lowes smoother) is approximately horizontal at zero.

- Assumption Two: Predictors (x) are Independent & Observed with Negligible Error.

The easiest way to check the assumption of independence is using the Durbin-Watson test. Running this test will give you an output with a p-value, which will help you determine whether the assumption is met or not.  

The null hypothesis states that the errors are not auto-correlated with themselves (they are independent). Thus, if we achieve a p-value > 0.05, we would fail to reject the null hypothesis. This would give us enough evidence to state that our independence assumption is met.

- Assumption Three: Residual Errors have a Mean Value of Zero.

We can easily check this assumption by looking at the same residual vs fitted plot. We would ideally want to see the red line flat on 0, which would indicate that the residual errors have a mean value of zero.

- Assumption Four: Residual Errors have Constant Variance.

We can check this assumption using the Scale-Location plot. In this plot we can look for the fitted values vs the square root of the standardized residuals. Ideally, we would want to see the residual points equally spread around the red line, which would indicate constant variance.

- Assumption Five: Residual Errors are Independent from Each Other & Predictors (x).

This assumption requires knowledge of study design or data collection in order to establish the validity of this assumption, so we will not be covering this in this blog.

#### Q4. Explain the concept of gradient descent. How is it used in machine learning?

Ans: Gradient descent is a iterative process. It is a first oder derivative optimization algorithm that is used to find the functions parameters (b0 and b1) that minimize a cost function to local minima or maxima. It is used to train machine learning models and neural networks. It tweaks its parameters iteratively to minimize a given function which is cost function to its local minimum.

The main objective of gradient descent is to minimize the cost function or the error between expected and actual. To minimize the cost function, two data points are required:

- Direction (forward and backward) & Learning Rate.

These two factors are used to determine the partial derivative calculation of future iteration and allow it to the point of local minimum or global minimum.

To achieve this goal, it performs two steps iteratively:

- Calculates the first-order derivative of the function to compute the gradient or slope of that function.

- Move away from the direction of the gradient, which means slope increased from the current point by alpha times, where Alpha is defined as Learning Rate. It is a tuning parameter in the optimization process which helps to decide the length of the steps.

In this sense it arrive at local minima or maxima with efficient parameters. Those parameters can then be used to predict our target variable.

#### Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?




Multicollinearity in multiple linear regression occurs when input variables are dependent on each other i.e they are correlated to each other. When there is correlated features in our dataset we can predict one feature using another but this is not a desired effect. It reduces model interpretability and affects results. 

Because of multicollinearity our model is unable to distinguish between particular effect of features on target variable. It gives insignificant statistical results. 

we can detect multicollinearity when : 
- Estimators have high variance and
- High sensitivity to small changes in input data. 

#### Q7. Describe the polynomial regression model. How is it different from linear regression?

Ans: Polynomial regression is a type of regression in which the relationship between the y target variable and x input variable is describes as the nth degree of polynomial. We use polynomial regression in case of non-linearity between features. In polynomial regression instead of a line fitting the data, we have a curve that fits the non-linear data. With increasing power of degree of polynomial our curve tries to fit data too well and leads to overfitting, so we create a balance in degree of features. 

Equation of polynomial regression is:

y = b0 + b1.x**1 + b2.x**2 + b2.x**3 +......+ bn.x**n

- Polynomial regression is different from other regression in terms that it used curve to fit data instead of a line in case of simple and multiple regression. 

- Also as it uses curve to fit the data that means there is non-linerity between features in dataset.

- In case of simple and multiple regression input features are raised to power of 1 degree of polynomial whereas in polynomial features are raised to power of nth degree according to use case. 

- we can interpret that the simple and multiple regression are a type of polynomial regression.
