#### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
print("""Just like we use Exception class as a base class for all the exceptions, in similar fashion we use lookuperror 
class for all the exceptions for error due to index error or key error in case of dictionary or a sequence.""")
print("")

x = [1,2,3,4,5]
try :
    x[6]
except IndexError as e:
    print("IndexError: x[6],", e)

print("Here we can see that indexing a number with wrong index will give us an index error.")
```
