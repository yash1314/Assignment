#### Q1. Explain GET and POST methods.

Ans - Get: The get method is used to request the servers to send some information or data. To use the GET method, we should have the following libraries installed like urllib,httplib, or requests. There are certain limits to getting data from server. In get method the data which we are sending is visible in url therefore it is not considered safe.

Post: The post method is used to send the data to the server. It is generally used when we want to uplaod some file or send some data like filling out a form, etc. There is no limit in post method for sending data.

Both the method can be used to send and recieve sensitive information therefore it should be handled with care.

#### Q2. Why is request used in Flask?

Ans - The request object in flask is used to handle the incoming data. The request object holds all the incoming data from the request. The flask request object is used to handle both sending and receiving operations from user browser to server and vice-versa.

#### Q3. Why is redirect() used in Flask?

Ans - A redirect is used in the Flask to send the user to a particular URL with the status code. When we access a website, our browser sends a request to the server, and the server replies with what is known as the HTTP status code, which is a three-digit number.

#### Q4. What are templates in Flask? Why is the render_template() function used?

Ans- Templates are files that contain static data as well as placeholders for dynamic data. A template is rendered with specific data to produce a final document. Flask uses the Jinja template library to render templates. In your application, you will use templates to render HTML which will display in the user's browser.

Flask provides us the render_template( ) function which can be used to render the external HTML file to be returned as the response from the view function.
