#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Ans: Scatter plot is a graph in which data points are represented in the form of dots. We have seen line graphs where data points join together and form a line, in scatter we only get dots which represents data points. We scatter plot we can understand the density of different data points, we can also understand outlier/ extreme values. We can create scatter plot using pandas, matplotlib, seaborn, etc.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = x, y = y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('X-Y graph')
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

Ans: With subplot() function we can draw multiplt plots in one figure. It adds subplots at a current figure using the grids position. It is useful in the sense that we can compare different attributes via figures side by side. 

```python
import numpy as np

# For line 1: 

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x,y)

# For line 2: 

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x,y)

# For line 3: 

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x,y)

# For line 4: 

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x,y)
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

Ans: Bar plot is a type of graph in which relationship between 2 variable is shown via heightened bars. The height of the bar represent the corresponding value on the y-axis. The bar graph can be vertical or horizontal in position. Bar plot is used to represent the relationship between different attributed.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit, width= 0.3)
```

```python
plt.barh(company, profit)
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Ans: Boxplots are a measure of how well distributed the data in a dataset is. It divides the dataset into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set. It is also useful in comparing the distribution of data across datasets by drawing boxplots for each of them.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot((box1, box2))
plt.grid(1)
```
