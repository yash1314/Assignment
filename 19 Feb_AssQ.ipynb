#### Q4.  Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:

    Company Name: ABC Corporation

    Location: India

    Contact Detail: 999-999-9999 

```python
# importing Flask from flask module
from flask import Flask

# creating the Flask object 
new_app = Flask(__name__)

# defining the project function.
@new_app.route('/')
def body():
    return """<h2>Company Name: ABC Corporation<br>
    Location: India<br>
    Contact Detail: 999-999-9999</h2>"""

# defining the welcome funciton.
@new_app.route("/welcome")
def welcome():
    return f"<h1>Welcome to ABC Corporation</h2> <br> {body()}"



if __name__ == "__main__":
    new_app.run(host="0.0.0.0", port = 8080)
```
